## SVM with hard margin

```python
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv('hard_margin_svm_data.csv')

array = data.to_numpy()

class1 = []
class2 = []

for p in array:
    if p[2] == 0:
        class1.append(p)
    else:
        class2.append(p)

plt.figure(figsize=(16, 8))
plt.scatter([p[0] for p in class1], [p[1] for p in class1], color='red')
plt.scatter([p[0] for p in class2], [p[1] for p in class2], color='blue')
plt.show()
```

```python
from sklearn.svm import SVC

data_x = array[:, :2]
data_y = array[:, 2]

classifier = SVC(C=1000.0, kernel='linear').fit(data_x, data_y)
```

```python
from sklearn.inspection import DecisionBoundaryDisplay

plt.figure(figsize=(16, 8))
plt.scatter([p[0] for p in class1], [p[1] for p in class1], color='red')
plt.scatter([p[0] for p in class2], [p[1] for p in class2], color='blue')

ax = plt.gca()
DecisionBoundaryDisplay.from_estimator(
    classifier,
    data_x,
    plot_method='contour',
    colors=['black', 'black', 'black'],
    levels=[-1, 0, 1],
    linestyles=['--', '-', '--'],
    ax=ax,
)

ax.scatter(
    classifier.support_vectors_[:, 0],
    classifier.support_vectors_[:, 1],
    s=200,
    linewidth=1,
    facecolors='none',
    edgecolors=['black'],
)

plt.show()
```

## KNN

```python
train_data = pd.read_csv('train_knn_data.csv')

train_array = train_data.to_numpy()

class1 = []
class2 = []
class3 = []

for p in train_array:
    if p[2] == 0:
        class1.append(p)
    elif p[2] == 1:
        class2.append(p)
    else:
        class3.append(p)

plt.figure(figsize=(16, 8))
plt.scatter([p[0] for p in class1], [p[1] for p in class1], color='red')
plt.scatter([p[0] for p in class2], [p[1] for p in class2], color='blue')
plt.scatter([p[0] for p in class3], [p[1] for p in class3], color='green')
plt.show()
```

```python
test_data = pd.read_csv('test_knn_data.csv')
test_array = test_data.to_numpy()
plt.figure(figsize=(16, 8))
plt.scatter(test_array[:, 0], test_array[:, 1], color='black')
plt.scatter([p[0] for p in class1], [p[1] for p in class1], color='red')
plt.scatter([p[0] for p in class2], [p[1] for p in class2], color='blue')
plt.scatter([p[0] for p in class3], [p[1] for p in class3], color='green')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5).fit(train_array[:, :2], train_array[:, 2])
```

```python
labels = classifier.predict(test_array[:, :2])

test_class1 = []
test_class2 = []
test_class3 = []

for i in range(len(test_array)):
    if labels[i] == 0:
        test_class1.append(test_array[i, :2])
    elif labels[i] == 1:
        test_class2.append(test_array[i, :2])
    else:
        test_class3.append(test_array[i, :2])

fig, axs = plt.subplots(1, 2, figsize=(16, 8))
axs[0].scatter([p[0] for p in class1], [p[1] for p in class1], color='red')
axs[0].scatter([p[0] for p in class2], [p[1] for p in class2], color='blue')
axs[0].scatter([p[0] for p in class3], [p[1] for p in class3], color='green')
axs[0].scatter(test_array[:, 0], test_array[:, 1], color='black')
axs[0].set_title('training labels')
axs[1].scatter([p[0] for p in test_class1], [p[1] for p in test_class1], color='red')
axs[1].scatter([p[0] for p in test_class2], [p[1] for p in test_class2], color='blue')
axs[1].scatter([p[0] for p in test_class3], [p[1] for p in test_class3], color='green')
axs[1].set_title('predicted labels')
plt.show()
```

## Decision Trees

```python
from sklearn import tree

data = pd.read_csv('diabetes.csv')
# shuffle data
data = data.sample(frac=1).reset_index()
data.head()
```

```python
training_data_size = int(len(data) * 0.7)

train_data = data.iloc[:training_data_size]
test_data = data.iloc[training_data_size:]

train_data_x = train_data.iloc[:, :-1]
train_data_y = train_data.iloc[:, -1]

test_data_x = test_data.iloc[:, :-1]
test_data_y = test_data.iloc[:, -1]
```

```python
classifier = tree.DecisionTreeClassifier(max_depth=4).fit(train_data_x, train_data_y)
```

```python
fig = plt.figure(figsize=(28, 8))
vis = tree.plot_tree(classifier,
                     feature_names=train_data_x.columns,
                     class_names=['Non-Diabetic', 'Diabetic'],
                     max_depth=4,
                     fontsize=9, 
                     proportion=True,
                     filled=True, 
                     rounded=True)
plt.show()
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import precision_recall_fscore_support

def print_metrics(classifier, data_x, data_y, threshold=0.5):
    prob_score = classifier.predict_proba(data_x)[:, 1]
    predicted_label = prob_score >= 0.5
    true_label = data_y >= 0.5
    acc = accuracy_score(true_label, predicted_label)
    print(f'accuracy (at threshold {threshold}): {acc}')

    confusion = confusion_matrix(true_label, predicted_label)
    print('')
    print('confusion matrix:')
    print(confusion)

    fnr = confusion[1, 0] / (confusion[0, 0] + confusion[1, 0])
    fpr = confusion[0, 1] / (confusion[0, 1] + confusion[1, 1])
    print('')
    print(f'FNR: {fnr}')
    print(f'FPR: {fpr}')

    roc_fpr, roc_tpr, roc_thresholds = roc_curve(data_y, prob_score)
    roc_auc = auc(roc_fpr, roc_tpr)

    print('')
    print(f'auc: {roc_auc}')

    precision, recall, fscore, _ = precision_recall_fscore_support(true_label, predicted_label, average='binary')

    print('')
    print(f'precision: {precision}')
    print(f'recall: {recall}')
    print(f'fscore: {fscore}')
```

```python
print('----------------')
print('training metrics')
print('----------------')
print_metrics(classifier, train_data_x, train_data_y)

print('')
print('------------')
print('test metrics')
print('------------')
print_metrics(classifier, test_data_x, test_data_y)
```

```python
classifier = tree.DecisionTreeClassifier().fit(train_data_x, train_data_y)
```

```python
print('----------------')
print('training metrics')
print('----------------')
print_metrics(classifier, train_data_x, train_data_y)

print('')
print('------------')
print('test metrics')
print('------------')
print_metrics(classifier, test_data_x, test_data_y)
```

## Bagging

```python
from sklearn.ensemble import BaggingClassifier
classifier = BaggingClassifier(n_estimators=10).fit(train_data_x, train_data_y)

print('----------------')
print('training metrics')
print('----------------')
print_metrics(classifier, train_data_x, train_data_y)

print('')
print('------------')
print('test metrics')
print('------------')
print_metrics(classifier, test_data_x, test_data_y)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=100, max_depth=4).fit(train_data_x, train_data_y)

print('----------------')
print('training metrics')
print('----------------')
print_metrics(classifier, train_data_x, train_data_y)

print('')
print('------------')
print('test metrics')
print('------------')
print_metrics(classifier, test_data_x, test_data_y)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

classifier = GradientBoostingClassifier(n_estimators=100, max_depth=3).fit(train_data_x, train_data_y)

print('----------------')
print('training metrics')
print('----------------')
print_metrics(classifier, train_data_x, train_data_y)

print('')
print('------------')
print('test metrics')
print('------------')
print_metrics(classifier, test_data_x, test_data_y)
```
